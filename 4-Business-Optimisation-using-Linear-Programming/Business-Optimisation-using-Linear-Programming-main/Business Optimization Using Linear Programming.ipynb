# Business Optimization Using Linear Programming with PuLP

### Project Description
This project demonstrates how optimization techniques can be used to solve business problems. Using Linear Programming and the PuLP library in Python, we determine the optimal production levels of two products to maximize profit while satisfying resource constraints.

Optimization techniques are widely used in business to allocate limited resources efficiently.
In this project, we solve a production planning problem using Linear Programming and PuLP.

### Business Problem

A manufacturing company produces two products.

The company wants to determine the optimal production quantity to maximize profit while considering:

Limited labor hours

Limited raw materials

### Mathematical Model

Decision variables:

A = units of Product A

B = units of Product B

Objective:

Maximize Profit = 20A + 30B

Subject to constraints:

2A+B≤40

A+3B≤60

A,B≥0

```python
import pulp
```

### Problem Statement

Scenario

Company: Smart Manufacturing Ltd.

They produce:

Product A → Electronic components

Product B → Mechanical components

Resources available:

| Resource     | Product A | Product B | Available |
| ------------ | --------- | --------- | --------- |
| Labor hours  | 2         | 1         | 40        |
| Raw material | 1         | 3         | 60        |


Profit:

Product A → $20 per unit

Product B → $30 per unit

Goal: Maximize total profit.

### Create Optimization Model

```python
# Create the Linear Programming model
model = pulp.LpProblem("Profit_Maximization", pulp.LpMaximize)
```

### Define Decision Variables

```python
# Decision variables
A = pulp.LpVariable('Product_A', lowBound=0)
B = pulp.LpVariable('Product_B', lowBound=0)
```

### Define Objective Function

```python
# Objective function
model += 20*A + 30*B
```

### Add Constraints

```python
# Labor constraint
model += 2*A + B <= 40

# Material constraint
model += A + 3*B <= 60
```

### Solve the Model

```python
model.solve()
```

### Print Results

```python
print("Status:", pulp.LpStatus[model.status])

print("Optimal Production Plan")

print("Units of Product A:", A.varValue)
print("Units of Product B:", B.varValue)

print("Maximum Profit:", pulp.value(model.objective))
```

### Business Insights

Key insights from the model:

The company should produce 12 units of Product A and 16 units of Product B.

This allocation fully utilizes available resources.

The maximum achievable profit is $720.

### Add Visualization of Constraints

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Plot the Constraints

```python
# Define x values
x = np.linspace(0, 50, 100)

# Constraints
y1 = 40 - 2*x        # 2A + B = 40
y2 = (60 - x) / 3    # A + 3B = 60

plt.figure()

plt.plot(x, y1, label="2A + B = 40")
plt.plot(x, y2, label="A + 3B = 60")

plt.xlim((0, 50))
plt.ylim((0, 50))

plt.xlabel("Product A")
plt.ylabel("Product B")
plt.title("Feasible Region for Production Optimization")

plt.legend()
plt.grid()

plt.show()
```
